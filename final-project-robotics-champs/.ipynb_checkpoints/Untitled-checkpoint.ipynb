```python
#import numpy, pandas, and matplotlib.pyplot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the file and create the DataFrame
col_names = ['Borough', 'Neighborhood', 'Building_Class_Category', 'Tax_Class_Current', 'Block', 'Lot', 'EaseMent', 'Building_Class_Current', 'Address', 'Apt._No.', 'Zip_Code', 'Residential_Units', 'Commercial_Units', 'Total_Units', 'Land_Sq_Ft', 'Gross_Sq_Ft', 'Year_Built', 'Tax_Class_AtSale', 'Building_Class_AtSale', 'Sale_Price', 'Sale_Date']
data=pd.read_csv("NYC_Housing_2018.csv", header=0, parse_dates=True, index_col='Sale_Date', names=col_names,  na_values=[' ', ' $-   '])
data.info()
data.head()
```

```python
#Delete the unnecessary columsn
data.drop(['EaseMent', 'Apt._No.'], axis=1, inplace=True)
data.dropna(axis=0, how='any', inplace=True)
#Change data types
data['Borough'] = data.Borough.astype(str)
data['Block'] = data.Block.astype(str)
data['Lot'] = data.Lot.astype(str)
data['Zip_Code'] = data.Zip_Code.astype(str)
data['Tax_Class_AtSale'] = data.Tax_Class_AtSale.astype(str)
data['Year_Built'] = data.Year_Built.astype(int)
data.info()
```

```python
#Find the shape of the data with ranges
#Describe the data
df= data[(data.Sale_Price >= 100000) & (data.Sale_Price <= 10000000)].copy()
df.shape
df.describe()
```

```python
#Plot the Residential Units of NYC Real Estate
df.plot(y='Residential_Units', use_index=True, color='blue')
plt.title('Trend: Residential Units')
plt.xlabel('Sale Date')
plt.ylabel('Number of Units')

#Plot the Commercial Units of Real Estate
df.plot(y='Commercial_Units', use_index=True, color='red')
plt.title('Trend: Commercial Units')
plt.xlabel('Sale Date')
plt.ylabel('Number of Units')

plt.show()
```

```python
#Plot histogram of Sale Price
plt.hist(df.Sale_Price, bins=300, histtype='stepfilled', alpha=0.7, color='green', rwidth=1)
plt.grid(axis='y', alpha=0.5)
plt.xlabel('Sale_Price')
plt.title('NYC 2018 Housing Sales Price')
plt.show()
```

```python
#Plot Bar Chart of Types of Neighborhoods with Sale Price
data['Neighborhood'].value_counts().plot.barh()
plt.ylabel('Neighborhood')
plt.xlabel('Sale_Price')
```

```python
#Plot Bar Chart of Types of Zip Code with Sale Price
data['Zip_Code'].value_counts().plot.barh()
plt.ylabel('Zip_Code')
plt.xlabel('Sale_Price')
```

```python
#Plot the Year built with the Sale Price
df.plot(y='Sale_Price', use_index=True, color='red')
plt.title('Sale Price with Year Built')
plt.xlabel('Year_Built')
plt.ylabel('Sale_Price')

plt.show()
```

```python
#Look at all of the different zip codes
data['Zip_Code'].value_counts()
```

```python
#Aggregate Zip Code
#How do i make into ranges for a pie chart instead
df.groupby([df.index.time,'Zip_Code'])['Zip_Code'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
#Plot the Commercial Units of Real Estate
df.plot(y='Commercial_Units', use_index=False, color='red')
plt.title('Commercial Units')
plt.xlabel('Gross_Sq_units')
plt.ylabel('Number of Units')

plt.show()
```
